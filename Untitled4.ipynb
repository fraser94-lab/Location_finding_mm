```python
import math

def distance(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # Convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(math.radians, [lon1, lat1, lon2, lat2])

    # Haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a)) 
    radius = 6371000 # Radius of earth in meters
    return c * radius
```

```python
import rasterio
from rasterio.mask import mask
from shapely.geometry import box

# Open the GeoTIFF file
with rasterio.open('3-2022-RichmondPark_modified.tif') as src:
    # Define the bounding box using minimum and maximum longitude and latitude values
    bbox = box(min_lon, min_lat, max_lon, max_lat)
    # Get the coordinates of the bounding box in the same CRS as the GeoTIFF file
    bbox_coords = [bbox]
    out_image, out_transform = mask(src, bbox_coords, crop=True)
    out_meta = src.meta.copy()

# Update the metadata of the output file with the new dimensions and transform
out_meta.update({'driver': 'GTiff',
                 'height': out_image.shape[1],
                 'width': out_image.shape[2],
                 'transform': out_transform})

# Write the output GeoTIFF file
with rasterio.open('3-2022-RichmondPark_modified_croped.tif', 'w', **out_meta) as dest:
    dest.write(out_image)
    
cropped = rasterio.open("3-2022-RichmondPark_modified_croped.tif")
show(cropped)
```

```python
import math

def haversine_distance(lon1, lat1, lon2, lat2):
    # Convert coordinates to radians
    lon1, lat1, lon2, lat2 = map(math.radians, [lon1, lat1, lon2, lat2])

    # Calculate differences between coordinates
    dlon = lon2 - lon1
    dlat = lat2 - lat1

    # Apply Haversine formula to calculate distance
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    r = 6371  # Radius of earth in kilometers
    distance = c * r

    return distance
```

```python
import rasterio
from rasterio.windows import Window

crop_file_path='/home/vboxuser/Documents/location_mm_data/crop_dataset_3'

# Open the input GeoTIFF file
with rasterio.open('3-2022-RichmondPark_modified_croped.tif') as src:
    
    # Get the size of the image in pixels
    width = src.width
    height = src.height
    
    # Define the size of the crop
    crop_width = int(width/10)
    crop_height = int(height/10)
    
    print('crop width is',crop_width)
    print('crop height is',crop_height)
    
    # Get the metadata of the input file
    meta = src.meta.copy()

    # Iterate over the windows
    for i, window in enumerate(src.block_windows(1)):
        
        print(window[0])

        # Get the window dimensions
        row_start, col_start, row_stop, col_stop = window[0]

        # Define the output file name
        output_file = os.path.join(crop_file_path,f'crop_{i}.tif')

        # Update the metadata of the output file
        meta.update({
            'height': row_stop - row_start,
            'width': col_stop - col_start,
            'transform': rasterio.windows.transform(window=((row_start, row_stop), (col_start, col_stop)), transform=meta['transform'])
        })

        # Read the data from the input file window
        data = src.read(window=window)

        # Write the cropped data to a new GeoTIFF file
        with rasterio.open(output_file, 'w', **meta) as dst:
            dst.write(data)

        print(f'Saved {output_file}')
```
